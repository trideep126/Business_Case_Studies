```python
#Renameing the Transaction amount coklumn
df.rename(columns={'TransactionAmount (INR)':'TransactionAmount'},inplace=True)
```

```python
df1 = df.copy()
```

```python
#handle date columns
df1['TransactionDate'] = pd.to_datetime(df1['TransactionDate'])
df1['CustomerDOB'] = pd.to_datetime(df1['CustomerDOB'])
```

```python
df1['TransactionDate']
```

```python
df1['TransactionTime'] = df1['TransactionTime'].astype(str).str.zfill(6)
df1['TransactionTime'] = pd.to_datetime(df1['TransactionTime'],format='%H%M%S').dt.time
df1['TransactionTime']
```

```python
df1.describe()
```

```python
df1.select_dtypes(include='object').describe()
```

* Around 1 lakh customers are from Mumbai
* Male is the predominant gender
* Customers born in 1800 cannot be alive today to make transactions in 2016
* Similarly, customers can't be born in the year 2073

```python
#find out customer's age
from datetime import datetime

today = datetime.today().year
df1['Age'] = today - df1['CustomerDOB'].dt.year
```

```python
df1.sample(10)
```

```python
#We're only keeping customers aged between 18 and 90
df2 = df1[df1['Age'].between(18,90)]
```

```python
#Check out how many transactions had invalid customer ages
print(df1.shape)
print(df2.shape)
```

```python
#Check if null values exist in the filtered dataframe
df2.isna().sum()
```

```python
#We choose to drop those null values, since they won't affect the data that much
df3 = df2.dropna()
print(df3.shape)
```

```python
#Check out the unique rows by transaction IDs
df3['TransactionID'].nunique()
```

```python
#Check out the unique rows by customer IDs
df3['CustomerID'].nunique()
```

```python
#Only keep rows where customer IDs are unique, drop the rest
df4 = df3.drop_duplicates(subset='CustomerID')
df4.shape
```

```python
#Drop unnecessary columns
df5 = df4.drop(['CustomerID','TransactionID','CustomerDOB'],axis=1)
df5.sample(3)
```

```python
#Change the data type of 'Age' column from float to integer
df5['Age'] = df5['Age'].astype('int64')
```

```python
df5.select_dtypes(include=object).describe()
```

Notice that there is clear imbalance in data since most individuals are from the male population.

```python
#Check out the count of individuals in each Gender
df5['CustGender'].value_counts()
```

```python
#Check out % of individuals in each Gender
df5['CustGender'].value_counts(normalize=True)*100
```

```python
#Show the number of population in each location
value_counts = df5['CustLocation'].value_counts()
value_counts_df = value_counts.reset_index()
value_counts_df.columns = ['CustLocation', 'ValueCounts']
value_counts_df
```

```python
#Locations with most frequent transactions
locations = value_counts_df[value_counts_df['ValueCounts'].between(1000,74796)]
locations_arr = list(locations['CustLocation'])
locations_arr
```

```python
#Considering only those transactions in popular locations

df6 = df5[df5['CustLocation'].isin(locations_arr)]
print(df5.shape)
print(df6.shape)
```
